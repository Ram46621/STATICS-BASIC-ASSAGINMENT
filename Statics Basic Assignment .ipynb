Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Answer:

Statistics can be broadly divided into two main categories: descriptive statistics and inferential statistics.

Descriptive statistics refers to methods used to summarize and describe the main features of a dataset. It provides simple summaries about the sample and the measures. These summaries may be numerical (mean, median, mode, standard deviation) or graphical (charts, histograms).

Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

Answer:

Sampling is the process of selecting a subset of individuals or observations from a large population to estimate characteristics of the whole population.

Since studying the entire population is often costly and time-consuming, sampling helps in faster and more efficient analysis.

Random Sampling

In random sampling, every individual in the population has an equal chance of being selected.

Example:
Picking 100 employees randomly from a company list.

Stratified Sampling

In stratified sampling, the population is divided into groups (called strata) based on certain characteristics such as age, gender, or department. Samples are then taken from each group.

Example:
Selecting employees from HR, Sales, and IT departments separately to ensure representation from each.

Difference:
Random sampling treats the population as one group, whereas stratified sampling ensures representation from different subgroups.

Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.

Answer:

Measures of central tendency describe the center or typical value of a dataset.

Mean:
The mean is the average of all values. It is calculated by dividing the sum of observations by the total number of observations.

Median:
The median is the middle value when the data is arranged in ascending or descending order.

Mode:
The mode is the value that appears most frequently in the dataset.

Importance

hese measures help in:

Understanding the overall pattern of data

Comparing different datasets

Supporting decision-making in business and research

For example, companies use averages to understand sales performance.



Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?

Answer:

Skewness measures the asymmetry of a probability distribution.

Positive skew (right-skewed): tail is longer on the right side.

Negative skew (left-skewed): tail is longer on the left side.

A positive skew implies that most values are concentrated on the lower side, but a few large values pull the average upward.

Example:
Income data often shows positive skew because a small number of people earn very high incomes.

Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers.


```python
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

import statistics as stats

mean_value = stats.mean(numbers)
median_value = stats.median(numbers)
mode_value = stats.mode(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)

```

Question 6: Compute covariance and correlation coefficient between two datasets.

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

covariance = np.cov(list_x, list_y)[0][1]
correlation = np.corrcoef(list_x, list_y)[0][1]

print("Covariance:", covariance)
print("Correlation:", correlation)
```

Question 8: Advertising spend vs sales relationship.

Explanation:
Covariance shows direction of relationship.
Correlation shows strength (between -1 and +1).

```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

correlation = np.corrcoef(advertising_spend, daily_sales)[0][1]
print("Correlation:", correlation)
```

Question 9: Survey distribution analysis.

We use:

Mean

Standard deviation

Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

print("Mean:", np.mean(survey_scores))
print("Std Dev:", np.std(survey_scores))

plt.hist(survey_scores)
plt.title("Survey Score Distribution")
plt.show()
```
